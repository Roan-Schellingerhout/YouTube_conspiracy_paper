```python
import pandas as pd
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv(r"../data/training_videos.csv")
df_2 = pd.read_csv(r"../data/channel_review.csv")
```

```python
all_words = " ".join(df["full_text"])
```

```python
word_counts = Counter(all_words.split())
```

```python
sort = pd.Series(word_counts).sort_values(ascending=False)
sort
```

```python
sort = sort[sort.index.str.len() > 1]
```

```python
sort.plot(loglog=True);
```

```python
sort.plot();
plt.xticks(rotation=70);
```
