```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ensemble = pd.read_csv("ML_results/ensembles.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
svm = pd.read_csv("ML_results/support-vector_machine.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
nn = pd.read_csv("ML_results/neural_network.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
ridge = pd.read_csv("ML_results/ridge_classification.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
logr = pd.read_csv("ML_results/logistic_regression.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
knn = pd.read_csv("ML_results/k-nearest_neighbors.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
```

```python
all_results = [(ensemble, "Ensemble"), (svm, "Support-vector machine"), 
               (nn, "Neural network"), (ridge, "Ridge regression"), 
               (logr, "Logistic regression"), (knn, "k-nearest neighbors")]

highscores = pd.DataFrame({name:result.iloc[0] 
                           for result, name in all_results}, 
                          index = ["Accuracy", "Precision", "Recall", "F1"]).T

highscores.plot(kind="bar", figsize = (16, 9))
plt.ylim(0.85, 0.975);

plt.yticks(fontsize=14)
plt.xticks(rotation=35, fontsize=14);

plt.ylabel("Score", fontsize=18);
plt.xlabel("Classifier", fontsize=18);

plt.legend(prop = {"size": 17.5});

plt.gcf().subplots_adjust(bottom=0.25)
plt.savefig('hyperparameters.pdf');
```
