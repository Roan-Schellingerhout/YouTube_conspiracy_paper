```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ensemble = pd.read_csv("ensembles.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
svm = pd.read_csv("support-vector_machine.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
nn = pd.read_csv("neural_network.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
ridge = pd.read_csv("ridge_classification.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
logr = pd.read_csv("logistic_regression.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
knn = pd.read_csv("k-nearest_neighbors.csv").sort_values("Accuracy", ascending=False).drop("Unnamed: 0", axis=1)
```

```python
all_results = [(ensemble, "Ensemble"), (svm, "Support-vector machine"), 
               (nn, "Neural network"), (logr, "Logistic regression"), 
               (ridge, "Ridge regression"), (knn, "k-nearest neighbors")]

highscores = pd.DataFrame({name:result.iloc[0] 
                           for result, name in all_results}, 
                          index = ["Accuracy", "Precision", "Recall", "F1"]).T

highscores.plot(kind="bar", figsize=(16, 9));
plt.ylim(0.85, 0.975);

plt.yticks(fontsize=14)
plt.xticks(rotation=45, fontsize=14);

plt.ylabel("Score", fontsize=18);
plt.xlabel("Classifier", fontsize=18);

plt.legend(prop = {"size": 15})
```
